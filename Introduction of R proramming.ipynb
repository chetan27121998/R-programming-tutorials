### What is R Programming 
    1. R is an interpreted computer programming language which was created by Ross Ihaka and Robert Gentleman at the        University of Auckland, New Zealand." 
    2. The R Development Core Team currently develops R. 
    3. It is also a software environment used to analyze statistical information, graphical representation, reporting, and  data modeling. 
    4. R is the implementation of the S programming language, which is combined with lexical scoping semantics.
