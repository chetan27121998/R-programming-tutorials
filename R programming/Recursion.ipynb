### WAP to find the factorial of number

```r
# First create a user defined function using loop
factorial <- function (num)
    {
    f= 1
    while(num>0)
        {
        f = f * num
        num = num - 1
    }
    return (f)
} 

# Main program
num = as.integer(readline("Enter the number : "))
# Call the function
cat("The factorial of given number is: ", factorial (num))
```

```r
# First create a user defined function using recursion
factorial <- function(n)
    {
    if(n==0)
        {
        return (1)
    }else
        {
        return (n*factorial(n-1))
    }
}
 

# Main program
n = as.integer(readline("Enter the number for factorial : "))
# Call the function
cat("The factorial of given number is : ",factorial(n))
```

```r
# Switch statement : it is also condition statement .
# It is use to create a menu driven program 
# for eg.
# 1. sum   2. substract  3. Multiply   4. divide
repeat
    {

a = as.integer(readline("Enter the number a : "))
b = as.integer(readline("Enter the number b : "))

cat("\n1. Sum \n2. Sunbstract \n3. Multiply \n4. Divide")
ch = as.integer(readline("Enter your choice : "))

switch(ch,cat("\nSum of 2 nos:",(a+b)),
cat("\nsubstract of 2 nos: ",(a-b)),
cat("\nMultiply of 2 nos : ",(a*b)),
cat("/nDivide of 2 nos : ", (a/b)),
cat("Invalid choice")
)

ans = readline("Continue ........Y/ N")
ans=tolower(ans)
if(ans=="n")
    {
    break  # Exit from Loop
}
    
    } # End of repeat Loop
```

```r
# 
aoc <- function() # No passing argument, Area of circle
    {
    r = as.integer(readline("Enter the radius : "))
    return (3.14*r*r)
}
aor<-function()  # No passing argument, Area of rectangle
    {
    l = as.integer(readline("Enter the length : "))
    w = as.integer(readline("Enter the width : "))
    return (l *w)
}
aot<-function() # No passing argument , area of triangle
    {
    b = as.integer(readline("Enter the Base : "))
    h = as.integer(readline("Enter the Height : "))
    return (b*h/2)
}
```

```r
# Main program
print("c. Area of circle : ")
print("r. Area of rectangle: ")
print("t. Area of Triangle: ")
ch = readline("Enter your choice: ")
ch = tolower(ch)
switch(ch,
      "c"=cat("Area of circle: ",aoc()),
      "r"=cat("Area of rectangle: ",aor()),
      "t"=cat("Area of triangle: ",aot()),
      cat("Invalid Choice")
      )
```
